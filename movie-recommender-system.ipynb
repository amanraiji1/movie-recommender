```python
import numpy as np
import pandas as pd
import ast #string to dictionary conversion
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
#merge both dataframes
movies = movies.merge(credits,on = 'title')
```

```python
movies.head().shape
```

```python
movies.info()
```

```python
#keeping only relevant features
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

# Preprocessing

```python
movies.isna().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.shape
```

```python
movies.duplicated().sum()
```

```python
movies['genres'][0]
```

```python
def convert(obj):
    L = []
    for x in ast.literal_eval(obj):
        L.append(x['name'])
    return L
```

## Example : ast to convert string dictionary to actual dictionary

```python
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
def fetch_top3_cast(obj):
    L = []
    count = 0
    for i in ast.literal_eval(obj):
        if count < 3:
            L.append(i['name'])
            count += 1
        else:
            break
    return L
        
```

```python
fetch_top3_cast(movies['cast'][0])
```

```python
movies['cast'] = movies['cast'].apply(fetch_top3_cast)
movies.head()
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies.head()
```

```python
#to remove space such that Sam Hidlle and Sam Thompson both are treated diffrently
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
# remove space
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
#split overview into words separated with ','
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_movies = movies.drop(columns = ['overview','genres','cast','crew','keywords'])
```

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x: " ".join(x))
new_movies.head()
```

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x: x.lower())
```

```python
new_movies.head()
```

```python
new_movies['tags'][0]
```

# Stemming

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(obj):
    y = []
    for i in obj.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
stem('loving love loved')
```

```python
new_movies['tags'] = new_movies['tags'].apply(stem)
```

```python
new_movies['tags'][0]
```

# Vectorization

```python
#we are going to use bagofwords technique for converting text to vectors
    #first we will merge all the tags into a single large sentence/textt
    #we will find most occuring 5000 words(any no. is fine) [w1,w2,w3......w5000]
    #then for each movie we will search how many times w1 appears,w2 appears....
        #this way we will create a row/vector like this [5 0 2 3..... 6]
    #now using this vector we will find similarity(cosine) between vectors
    #select top 5 and display those
#[Note that we will not include stopwords like is,am,are,.... ]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vector = cv.fit_transform(new_movies['tags']).toarray()
```

```python
vector.shape
```

```python
cv.get_feature_names_out()
```

# Cosine Similarity 

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity[0]
```

```python
similarity.shape
```

# Recommenndation function

```python
sorted(similarity[0],reverse=True)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x : x[1])
```

```python
def recommend(movie):
    index = new_movies[new_movies['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_movies.iloc[i[0]].title)
```

```python
recommend('Batman')
```
